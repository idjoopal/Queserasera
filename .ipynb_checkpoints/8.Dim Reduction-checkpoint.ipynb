# 8. 차원축소

- 차원축소는 정보가 유실되기 때문에 정말 축소가 필요한지 확인해야한다.

## 8.1 차원의 저주
- 고차원 데이터 셋은 예측을 위해 훨씬 더 많은 외삽 extrapolation을 해야하므로 예측이 불안정하며, 과대적합 위험이 커진다.

## 8.2 차원 축소를 위한 접근 방법

### 8.2.1 투영
- ex) 3차원의 데이터를 어떤 2차원 평면에 수직이 되도록 투영하여 2차원화 한다.

### 8.2.2 매니폴드 학습
- 스위스롤은 2D매니폴드의 예시
- 매니폴드 가정 : 고차원 데이터셋이 더 낮은 저차원 매니폴드에 가깝게 놓여있다.

## 8.3 PCA
- 데이터에 가장 가까운 초평면을 정의 한 다음, 데이터를 이 평면에 투영함

### 8.3.1 분산보존
- 분산이 가장 크게 보존되는 축을 선택하는 것이 가장 좋은 투영방법
- 원본과 투영된 것의 평균 제곱거리가 최소인 것.

### 8.3.2 PCA
- 이런 축을 선택하는 것.
- n차원 데이터에서 하나씩 축을 선택해 나가는 것.
- i번째 축을 i번째 주성분이라 한다.
- 특잇값분해(SVD)를 사용하여 표준 행렬 분해를 하면 세개 행렬 곱셈인 $U\sum_{}^{} V^T$ 로 분해할 수 있다.

```python
np.random.seed(4)
m = 60
w1, w2 = 0.1, 0.3
noise = 0.1

angles = np.random.rand(m) * 3 * np.pi / 2 - 0.5
X = np.empty((m, 3))
X[:, 0] = np.cos(angles) + np.sin(angles)/2 + noise * np.random.randn(m) / 2
X[:, 1] = np.sin(angles) * 0.7 + noise * np.random.randn(m) / 2
X[:, 2] = X[:, 0] * w1 + X[:, 1] * w2 + noise * np.random.randn(m)

X_centered = X - X.mean(axis=0)
U, s, Vt = np.linalg.svd(X_centered)
c1 = Vt.T[:, 0]
c2 = Vt.T[:, 1]
```

### 8.3.3 d차원 투영
- 주성분을 추출하여 d개의 초평면에 투영하면 d차원으로 축소할 수 있다.
- d차원으로 축소된 데이터셋X = 행렬X 와 V의 첫 d열로 구성된 W_d를 행렬곱셈
- $X_{d-proj} = XW_d$

```python
m, n = X.shape

S = np.zeros(X_centered.shape)
S[:n, :n] = np.diag(s)

np.allclose(X_centered, U.dot(S).dot(Vt))

# 첫 두개의 주성분
W2 = Vt.T[:, :2]
X2D = X_centered.dot(W2)

X2D_using_svd = X2D
```

### 8.3.4 사이킷런 사용
- SVD분해를 사용하여 PCA를 구현한다.

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X2D = pca.fit_transform(X)
```

## 8.4 커널 PCA

- SVM에서의 커널트릭 기법을 사용하여 비선형 투영을 진행하는 것.
- 샘플의 군집을 유지하거나 꼬인 매니폴드를 펼침(스위스 롤 같은 것)
- 사이킷런의 kernelPCA를 사용한다.

```python
# 스위스롤 데이터 활용
from sklearn.datasets import make_swiss_roll
X, t = make_swiss_roll(n_samples=1000, noise=0.2, random_state=42)

# 가우시안 rbf 커널을 사용하는 예시
from sklearn.decomposition import KernelPCA
rbf_pca = KernelPCA(n_components=2, kernel="rbf", gamma=0.04)
X_reduced = rbf_pca.fit_transform(X)
```

```python
# 스위스롤 데이터셋을 여러가지 커널을 활용하여 PCA 진행하기
lin_pca = KernelPCA(n_components=2, kernel="linear", fit_inverse_transform=True)
rbf_pca = KernelPCA(n_components=2, kernel="rbf", gamma=0.0433, fit_inverse_transform=True)
sig_pca = KernelPCA(n_components=2, kernel="sigmoid", gamma=0.001, coef0=1, fit_inverse_transform=True)

#y = t > 6.9

plt.figure(figsize=(11, 4))
for subplot, pca, title in ((131, lin_pca, "Linear kernel"), (132, rbf_pca, "RBF kernel, $\gamma=0.04$"), (133, sig_pca, "Sigmoid kernel, $\gamma=10^{-3}, r=1$")):
    X_reduced = pca.fit_transform(X)
    if subplot == 132:
        X_reduced_rbf = X_reduced
    
    plt.subplot(subplot)
    #plt.plot(X_reduced[y, 0], X_reduced[y, 1], "gs")
    #plt.plot(X_reduced[~y, 0], X_reduced[~y, 1], "y^")
    plt.title(title, fontsize=14)
    plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=t, cmap=plt.cm.hot)
    plt.xlabel("$z_1$", fontsize=18)
    if subplot == 131:
        plt.ylabel("$z_2$", fontsize=18, rotation=0)
    plt.grid(True)

#save_fig("kernel_pca_plot")
plt.show()
```

### 8.4.1 커널 선택과 Hyperparameter 튜닝
- 그리드 서치를 진행하여 좋은 커널과 파라미터를 선택할 수 있다.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

clf = Pipeline([
        ("kpca", KernelPCA(n_components=2)),
        ("log_reg", LogisticRegression())
    ])

param_grid = [{
        "kpca__gamma": np.linspace(0.03, 0.05, 10),
        "kpca__kernel": ["rbf", "sigmoid"]
    }]

grid_search = GridSearchCV(clf, param_grid, cv=3)
grid_search.fit(X, y)
```

```python
print(grid_search.best_params_)
```

- 비지도 학습 방법으로 재구성 오차가 가장 적은 커널과 파라미터를 선택할 수 있다.

```python
rbf_pca = KernelPCA(n_components=2, kernel="rbf", gamma=0.0433,
                    fit_inverse_transform=True)
X_reduced = rbf_pca.fit_transform(X)
X_preimage = rbf_pca.inverse_transform(X_reduced)

from sklearn.metrics import mean_squared_error
mean_squared_error(X, X_preimage)
# 이 값이 최소화 될 수있는 커널과 감마값을 찾는다.
# 그리드서치와 교차검증한다.
```

```python
plt.figure(figsize=(20, 6))
plt.subplot(132)
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=t, cmap=plt.cm.hot, marker="x")
plt.xlabel("$z_1$", fontsize=18)
plt.ylabel("$z_2$", fontsize=18, rotation=0)
plt.grid(True)
```

## 8.5 LLE
- locally linear embedding(LLE) 지역선형임베딩
- 비선형 차원축소의 기술이다.
- 투영에 의존하지 않는다.
1. 데이터간 이웃한 것을 찾아 선형적으로 연관되어있는지 측정
2. 이웃 끼리 묶어 그 관계가 잘 보존되는 저차원을 찾는다.

```python
X, t = make_swiss_roll(n_samples=1000, noise=0.2, random_state=41)

from sklearn.manifold import LocallyLinearEmbedding

# n_neighbors 값은 범주내의 데이터 개수보다 쪼끔 더 많이 주는 것이 좋다
# if you know the classes of your samples, you'd better set n_neighbors a little bit greater than the number of samples in each class. 
lle = LocallyLinearEmbedding(n_components=2, n_neighbors=10, random_state=42)
X_reduced = lle.fit_transform(X)

plt.title("Unrolled swiss roll using LLE", fontsize=14)
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=t, cmap=plt.cm.hot)
plt.xlabel("$z_1$", fontsize=18)
plt.ylabel("$z_2$", fontsize=18)
plt.axis([-0.065, 0.055, -0.1, 0.12])
plt.grid(True)

plt.show()
```

## 8.6 다른 차원 축소 기법
- 랜덤투영
    - 랜덤한 선형투영을 사용해 저차원 공간으로 투영함

- 다차원 스케일링(MDS)
    - 샘플간 거리를 보존하며 차원 축소
    
- Isomap
    - 각 샘플을 가장 가까운 이웃과 연결하는 식으로 축소
    
- t-SNE
    - 비슷한 것은 가깝게, 비슷하지 않은 것은 더 멀어지게 축소
    - 시각화에 주로 사용
    
- 선형 판별분석(LDA)
    - 분류 알고리즘이지만
    - 범주 구분을 가장 잘하는 축을 학습함.

```python
from sklearn.manifold import MDS

mds = MDS(n_components=2, random_state=42)
X_reduced_mds = mds.fit_transform(X)

from sklearn.manifold import Isomap

isomap = Isomap(n_components=2)
X_reduced_isomap = isomap.fit_transform(X)

from sklearn.manifold import TSNE

tsne = TSNE(n_components=2, random_state=42)
X_reduced_tsne = tsne.fit_transform(X)

from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

lda = LinearDiscriminantAnalysis(n_components=2)
X_mnist = mnist["data"]
y_mnist = mnist["target"]
lda.fit(X_mnist, y_mnist)
X_reduced_lda = lda.transform(X_mnist)

titles = ["MDS", "Isomap", "t-SNE"]

plt.figure(figsize=(11,4))

for subplot, title, X_reduced in zip((131, 132, 133), titles,
                                     (X_reduced_mds, X_reduced_isomap, X_reduced_tsne)):
    plt.subplot(subplot)
    plt.title(title, fontsize=14)
    plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=t, cmap=plt.cm.hot)
    plt.xlabel("$z_1$", fontsize=18)
    if subplot == 131:
        plt.ylabel("$z_2$", fontsize=18, rotation=0)
    plt.grid(True)


plt.show()
```
