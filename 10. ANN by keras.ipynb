### 10.1.3 퍼셉트론(TLU)
- 각각의 입력의 가중치 합을 계산 한 뒤, 그 합에 계단 함수를 적용하여 결과를 출력
- 헤비사이드(heaviside step function) 계단 함수를 주로 사용한다. (heavy가 아닌 발명자 이름이다.)
    - z<0 일때 0, z>=0 일때 1
- 부호함수를 사용하기도 한다.
    - z<0 일때 -1, z=0 일때 0, z>0 일때 1
- 하나의 TLU는 선형 이진 분류에 사용할 수 있다.

```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.linear_model import Perceptron

iris = load_iris()
X = iris.data[:, (2, 3)]  # 꽃잎 길이, 꽃잎 너비
y = (iris.target == 0).astype(np.int)

per_clf = Perceptron(max_iter=1000, tol=1e-3, random_state=42)
# per_clf = Perceptron()
# 아무것도 입력안하면 parameter가 default로 적용됨.
per_clf.fit(X, y)

y_pred = per_clf.predict([[2, 0.5]])
print(y_pred)
```

### 10.1.4 다층 퍼셉트론과 역전파
- 단층 퍼셉트론은 XOR을 해결할 수 없다.
- 퍼셉트론을 여러층 쌓아 올리면 일부 문제를 해결할 수 있다.
- 입력과 출력 사이에 은닉층을 여러개 쌓아올려 심층 신경망이라 부른다.
- 역전파 훈련 알고리즘을 통해 훈련을 진행한다.
    - 이 알고리즘의 원활한 작동을 위해 계단함수를 로지스틱(시그모이드) 함수로 바꾼다.
    - 계산할 그라디언트를 주어서 경사 하강법에 사용할 수 있다.
    - Sigmoid 함수
    - tanh(하이퍼볼릭 탄젠트 함수)
    - ReLU

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def relu(z):
    return np.maximum(0, z)

def derivative(f, z, eps=0.000001):
    return (f(z + eps) - f(z - eps))/(2 * eps)

z = np.linspace(-5, 5, 200)

plt.figure(figsize=(11,4))

plt.subplot(121)
plt.plot(z, np.sign(z), "r-", linewidth=1, label="Step")
plt.plot(z, sigmoid(z), "g--", linewidth=2, label="Sigmoid")
plt.plot(z, np.tanh(z), "b-", linewidth=2, label="Tanh")
plt.plot(z, relu(z), "m-.", linewidth=2, label="ReLU")
plt.grid(True)
plt.legend(loc="center right", fontsize=14)
plt.title("Activation functions", fontsize=14)
plt.axis([-5, 5, -1.2, 1.2])

plt.subplot(122)
plt.plot(z, derivative(np.sign, z), "r-", linewidth=1, label="Step")
plt.plot(z, derivative(sigmoid, z), "g--", linewidth=2, label="Sigmoid")
plt.plot(z, derivative(np.tanh, z), "b-", linewidth=2, label="Tanh")
plt.plot(z, derivative(relu, z), "m-.", linewidth=2, label="ReLU")
plt.grid(True)
#plt.legend(loc="center right", fontsize=14)
plt.title("Derivatives", fontsize=14)
plt.axis([-5, 5, -0.2, 1.2])

plt.show()
```

### 10.2.5 서브클래싱 API로 동적 모델 만들기
- 시퀀셜과 함수형 API는 미리 선언함
    - 저장, 복사, 공유가 용이함
    - 모델의 구조를 출력하고 분석하기 좋음
    - 모델 구조만 보고 에러를 일찍 발견할 수 있음
    - 하지만 좀더 동적인 구조가 필요할 수 있다.

```python
class WideAndDeepModel(keras.models.Model):
    def __init__(self, units=30, activation="relu", **kwargs):
        super().__init__(**kwargs)
        self.hidden1 = keras.layers.Dense(units, activation=activation)
        self.hidden2 = keras.layers.Dense(units, activation=activation)
        self.main_output = keras.layers.Dense(1)
        self.aux_output = keras.layers.Dense(1)
        
    def call(self, inputs):
        input_A, input_B = inputs
        hidden1 = self.hidden1(input_B)
        hidden2 = self.hidden2(hidden1)
        concat = keras.layers.concatenate([input_A, hidden2])
        main_output = self.main_output(concat)
        aux_output = self.aux_output(hidden2)
        return main_output, aux_output

model = WideAndDeepModel(30, activation="relu")
```

- call 안에서 다양한 조건문을 활용한 다양한 모델을 생성할 수 있음
- call 함수에 숨겨진 상태이므로 모델을 저장하거나 복사할 수 없음
- summary를 사용할 때 층간 연결 정보를 얻을 수 없음

### 10.2.7 콜백 사용하기
- 콜백을 사용하여 훈련 중간에 객체를 호출할 수있다.
- ModelCheckPoint를 사용하여 매 epoch마다 모델을 저장하고, Val점수가 가장 높을 때를 저장하도록 할 수 있다.

```python
keras.backend.clear_session()
np.random.seed(42)
tf.random.set_seed(42)

model = keras.models.Sequential([
    keras.layers.Dense(30, activation="relu", input_shape=[8]),
    keras.layers.Dense(30, activation="relu"),
    keras.layers.Dense(1)
])

model.compile(loss="mse", optimizer=keras.optimizers.SGD(lr=1e-3))
checkpoint_cb = keras.callbacks.ModelCheckpoint("my_keras_model.h5", save_best_only=True)
history = model.fit(X_train, y_train, epochs=10,
                    validation_data=(X_valid, y_valid),
                    callbacks=[checkpoint_cb])
model = keras.models.load_model("my_keras_model.h5") # 최상의 모델로 롤백
mse_test = model.evaluate(X_test, y_test)
```

```python
model.compile(loss="mse", optimizer=keras.optimizers.SGD(lr=1e-3))
early_stopping_cb = keras.callbacks.EarlyStopping(patience=10,
                                                  restore_best_weights=True)
history = model.fit(X_train, y_train, epochs=100,
                    validation_data=(X_valid, y_valid),
                    callbacks=[checkpoint_cb, early_stopping_cb])
mse_test = model.evaluate(X_test, y_test)
```

```python
# 훈련에 사용한다면 on_train, on_epoch, on_batch를 사용한다.
# 평가evaluate에 사용한다면 on_test, on_test_batch를 사용한다.
# 예측predict에 사용한다면 on_predict, on_predict_batch를 사용한다.
# _start, _end를 붙여 언제 쓸건지 선택한다.

# epoch의 끝에 검증과 훈련 손실을 출력한다.
class PrintValTrainRatioCallback(keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs):
        print("\nval/train: {:.2f}".format(logs["val_loss"] / logs["loss"]))
        

val_train_ratio_cb = PrintValTrainRatioCallback()
history = model.fit(X_train, y_train, epochs=3,
                    validation_data=(X_valid, y_valid),
                    callbacks=[val_train_ratio_cb])
```

```python
keras.backend.clear_session()
np.random.seed(42)
tf.random.set_seed(42)

# 케라스 모델을 사이킷런처럼 쓸수 있도록 만들기
def build_model(n_hidden=1, n_neurons=30, learning_rate=3e-3, input_shape=[8]):
    model = keras.models.Sequential()
    model.add(keras.layers.InputLayer(input_shape=input_shape))
    for layer in range(n_hidden):
        model.add(keras.layers.Dense(n_neurons, activation="relu"))
    model.add(keras.layers.Dense(1))
    optimizer = keras.optimizers.SGD(lr=learning_rate)
    model.compile(loss="mse", optimizer=optimizer)
    return model

keras_reg = keras.wrappers.scikit_learn.KerasRegressor(build_model)
keras_reg.fit(X_train, y_train, epochs=100,
              validation_data=(X_valid, y_valid),
              callbacks=[keras.callbacks.EarlyStopping(patience=10)])
```

```python
from scipy.stats import reciprocal
from sklearn.model_selection import RandomizedSearchCV

param_distribs = {
    "n_hidden": [0, 1, 2, 3],
    "n_neurons": np.arange(1, 100).tolist(),
    "learning_rate": reciprocal(3e-4, 3e-2).rvs(1000).tolist(),
}

rnd_search_cv = RandomizedSearchCV(keras_reg, param_distribs, n_iter=10, cv=3, verbose=2)
rnd_search_cv.fit(X_train, y_train, epochs=100,
                  validation_data=(X_valid, y_valid),
                  callbacks=[keras.callbacks.EarlyStopping(patience=10)])
```

### 10.3.1 은닉층 개수
- 1, 2개의 은닉층으로도 문제는 해결 잘된다.
- 복잡할 수록 은닉층을 점진적으로 추가한다.

### 10.3.2 은닉층의 뉴런 개수
- 입력과 출력층은 데이터와 출력결과에 의해 정해짐
- 일반적으로는 점점 숫자를 줄여나가는 방식
- 요즘엔 꼭 그런건 아니고 보통은 첫 은닉층이 큰 것이 낫다.
- 많은 뉴런을 주고, 과대적합을 방지하는 조기 종료나 규제 기법을 사용하는 것이 효과적.(스트레치 팬츠 방식)

### 10.3.3 학습률, 배치크기, 하이퍼파라미터 등
- 학습률
    - 최적 학습률은 훈련 알고리즘이 발산하는 최대 학습률의 절반정도.
    - 낮은 것부터 점진적으로 매우 높은 값까지 수백번 반복하는 것이 효과적
- 옵티마이저
    - 고전적 미니배치 경사하강법보다 좋은 것을 선택하는 것...
- 배치크기
    - 큰 배치를 사용하여 GPU를 효율적으로 활용하고 빠른결과를 내는 것이 좋지만
    - 훈련 초에 불안정하고 일반화 성능이 떨어지는 단점이 있다.
    - 작은 학습률부터 점차 올리는 학습률 예열 기법을 사용하여 이러한 단점을 극복할 수 있지만
    - 그래도 불안정하거나 성능이 만족스럽지 못하면 작은 배치(2~32)를 사용
- 활성화 함수
    - 일반적으로 ReLU가 가장 좋은 기본값이다.
- 반복횟수
    - 튜닝할 필요없이 조기종료 하는 방식을 사용한다.
